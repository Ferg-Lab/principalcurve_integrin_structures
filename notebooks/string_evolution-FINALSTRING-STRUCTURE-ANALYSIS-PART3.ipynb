## Format for plotting distances

```python
# Transposing the dictionary of dictionaries
transposed_exp_distances_domains_string = {}
for outer_key, inner_dict in exp_distances_domains_string.items():
    for inner_key, value in inner_dict.items():
        # Create the inner dictionary if it doesn't exist
        if inner_key not in transposed_exp_distances_domains_string:
            transposed_exp_distances_domains_string[inner_key] = {}
        # Swap the outer and inner keys
        transposed_exp_distances_domains_string[inner_key][outer_key] = value
        
transposed_string_distances_domains_string = {}
for outer_key, inner_dict in string_distances_domains_string.items():
    for inner_key, value in inner_dict.items():
        # Create the inner dictionary if it doesn't exist
        if inner_key not in transposed_string_distances_domains_string:
            transposed_string_distances_domains_string[inner_key] = {}
        # Swap the outer and inner keys
        transposed_string_distances_domains_string[inner_key][outer_key] = value
```

```python
domain_labels = {
    'beta_propeller_com': r'$\beta$-Propeller',
    'thigh_com': r'Thigh',
    'calf_1_com': r'Calf-1',
    'calf_2_com': r'Calf-2',
    'alpha_transmembrane_com': r'$\alpha$-TM',
    'beta_1_com': r'$\beta_i$',
    'hyb_com': r'Hyb',
    'psi_com': r'Psi',
    'e1_com': r'$E_1$',
    'e2_com': r'$E_2$',
    'e3_com': r'$E_3$',
    'e4_com': r'$E_4$',
    'beta_t_com': r'$\beta_T$',
    'beta_transmembrane_com': r'$\beta$-TM'
}

exp_xlabels_string = {}
exp_xvals_string = {}
exp_yvals_string = {}
exp_dict_xlabels_string = {}


string_xlabels_string = {}
string_xvals_string = {}
string_yvals_string = {}
string_dict_xlabels_string = {}


for k in ['exp', 'string']:
    
    if k == 'exp':
        use_dict = transposed_exp_distances_domains_string
    else:
        use_dict = transposed_string_distances_domains_string
        
    for string_number, dict_pdbs  in use_dict.items():

        xlabels_string_pdb = {}
        xvals_string_pdb = {}
        yvals_string_pdb = {}
        xlabels_dict_string_pdb = {}

        for pdb_name, dict_dict_distances in dict_pdbs.items():

            xlabels = []
            xvals = []
            yvals = []            
            xlabels_based_dict = {}
            
            for dist_dist_keys, dict_distances in dict_dict_distances.items():

                for dist_keys, distances in dict_distances.items():

                    # remove duplicate pair combinations
                    if dist_keys+':'+dist_dist_keys not in xvals:

                        xlabels.append(domain_labels[dist_dist_keys]+':'+domain_labels[dist_keys])

                        xvals.append(dist_dist_keys+':'+dist_keys)
                        yvals.append(distances/10)
                        
                        xlabels_based_dict[domain_labels[dist_dist_keys]+':'+domain_labels[dist_keys]] = distances/10

                        
            xlabels_string_pdb[pdb_name] = xlabels
            xvals_string_pdb[pdb_name] = xvals
            yvals_string_pdb[pdb_name] = yvals
            xlabels_dict_string_pdb[pdb_name] = xlabels_based_dict

        if k == 'exp':
            exp_xlabels_string[string_number] = xlabels_string_pdb
            exp_xvals_string[string_number] = xvals_string_pdb
            exp_yvals_string[string_number] = yvals_string_pdb
            exp_dict_xlabels_string[string_number] = xlabels_dict_string_pdb
        else:
            string_xlabels_string[string_number] = xlabels_string_pdb
            string_xvals_string[string_number] = xvals_string_pdb
            string_yvals_string[string_number] = yvals_string_pdb
            string_dict_xlabels_string[string_number] = xlabels_dict_string_pdb
```
